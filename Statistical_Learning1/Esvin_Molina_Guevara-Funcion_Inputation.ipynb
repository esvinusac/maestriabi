## Inputation
Notebook con el código en Python (una función) para hacer inputation y la realización de pruebas con tres datasets.

Los datasets utilizados son:
- cardio_train.csv
- diabetes.csv
- train_HP.csv

```python
# Autor: Esvin Molina Guevara.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def inputationColumna(dataset, nombreColumnaInputar, nombreColumnaTarget):
    #obtener la proporción de NaN que hay en la columna
    colPropNaN = dataset[nombreColumnaInputar].isnull().mean() 

    #si la proporción de NaN en la columna es mayor al 50% entonces 
    #esa columna es candidata a eliminarla
    if colPropNaN > 0.5:
        print('La columna tiene mas del 50% de faltantes.  Se recomienda eliminarla del analisis.')    
        return

    #calculamos la media de los valores de la columna a inputar
    meanValue = np.round(dataset[nombreColumnaInputar].mean(), 0)

    #calculamos la mediana de los valores de la columna a inputar
    medianValue = np.round(dataset[nombreColumnaInputar].median(), 0)

    #inputamos los valores NaN con la  media
    dataset_Col_meanInp = dataset[nombreColumnaInputar].fillna(meanValue)

    #imputamos los valores NaN con la  mediana
    dataset_Col_medianInp = dataset[nombreColumnaInputar].fillna(medianValue)

    #calculamos al correlacion entre la variable dependiente y la variable target,
    #tanto de las inputations con la media y con la mediana y asi ver cual es la 
    #que debemos tomar en cuenta
    meanCorr = np.corrcoef(dataset_Col_meanInp, dataset[nombreColumnaTarget])[0,1]
    medianCorr = np.corrcoef(dataset_Col_medianInp, dataset[nombreColumnaTarget])[0,1]
    
    print('Proporcion de NaN en la columna previo a la inputation: ', dataset[nombreColumnaInputar].isnull().mean() )

    if meanCorr == medianCorr:
        print('La inputation por la media o mediana son iguales.  Hacemos inputation con la media.')    
        dataset[nombreColumnaInputar] = dataset[nombreColumnaInputar].fillna(meanValue)
    elif meanCorr > medianCorr:
        print('Hacemos inputation con la MEDIA debido a que tiene una mayor correlacion con la columna target.')
        dataset[nombreColumnaInputar] = dataset[nombreColumnaInputar].fillna(meanValue)
    else:
        print('Hacemos inputation con la MEDIANA debido a que tiene una mayor correlacion con la columna target.')
        dataset[nombreColumnaInputar] = dataset[nombreColumnaInputar].fillna(medianValue) 
        
    print('Proporcion de NaN en la columna despues de hacer la inputation: ', dataset[nombreColumnaInputar].isnull().mean() )
    print('Valores de la columna despues de hacer la inputation:\n',dataset[nombreColumnaInputar])         
    
    
print('\nInputation con el primer dataset:')
print('**********************************')
# cargamos el dataset de los datos a usar
ds1 = pd.read_csv('train_HP.csv')
ds1ColumnaInputar = 'LotFrontage'
ds1ColumnaTarget = 'SalePrice'
inputationColumna(dataset = ds1, nombreColumnaInputar = ds1ColumnaInputar, nombreColumnaTarget = ds1ColumnaTarget)

print('\nInputation con el segundo dataset:')
print('**********************************')
# cargamos el dataset de los datos a usar
ds2 = pd.read_csv('diabetes.csv')
ds2ColumnaInputar = 'Glucose'
ds2ColumnaTarget = 'Outcome'
inputationColumna(dataset = ds2, nombreColumnaInputar = ds2ColumnaInputar, nombreColumnaTarget = ds2ColumnaTarget)

print('\nInputation con el tercer dataset:')
print('**********************************')
# cargamos el dataset de los datos a usar
ds3 = pd.read_csv('cardio_train.csv')
ds3ColumnaInputar = 'weight'
ds3ColumnaTarget = 'cardio'
inputationColumna(dataset = ds3, nombreColumnaInputar = ds3ColumnaInputar, nombreColumnaTarget = ds3ColumnaTarget)
```
