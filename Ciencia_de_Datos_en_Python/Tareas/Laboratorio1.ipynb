# Git

Git es un sistema de control de versiones distribuido. Esto significa que cada miembro del equipo tiene una copia local que contiene todo el código fuente de un proyecto, incluido el historial completo de todos los cambios de todos los archivos. Esta copia se llama "**repositorio**" y se encuentra en una carpeta oculta ".git" en el sistema de archivos local.

## Instalar Git 
Si queremos comenzar a utilizar Git como programa de control de versiones, en primer lugar, debemos familiarizarnos con el propio software y su interfaz de usuario. 

Git está disponible para Windows, Unix/Linux y macOS, con diferentes versiones que presentan pequeñas diferencias de uso. Después de llevar a cabo la instalación estándar correspondiente, podemos controlar el programa con el símbolo del sistema o una interfaz gráfica de usuario en cualquiera de los sistemas operativos.

En la página oficial del proyecto [Git](https://git-scm.com/), podemos encontrar los archivos de instalación binarios, las instrucciones para instalar el administrador de paquetes (sistemas Unix) y las ediciones portátiles listas para usar para cada sistema operativo. Solamente debemos descargar el paquete de instalación deseado o elegir el paquete apropiado mediante el administrador y, a continuación, seguir las instrucciones del asistente. 

## Crear o clonar un repositorio Git
El repositorio Git es el directorio central de un proyecto y también el principal punto de contacto para todos los participantes, a través del cual se lleva a cabo el control de todas las versiones. Por ello, el primer paso consiste en crear este repositorio principal o clonarlo (en forma de copia de trabajo) en caso de que vayamos a incorporarnos a un proyecto que ya esté siendo gestionado con Git.

Ya sea si queremos configurar el control de versiones para un proyecto nuevo o si recien instalamos Git,  para aprender a trabajar con la herramienta debemos crear un repositorio nuevo. Para ello, ejecutamos el comando *cd (change directory)* para acceder al directorio de nuestro dispositivo donde queremos ubicarlo:

```
# Nos ubicamos en el directorio donde queremos crear el nuevo repositorio
cd ruta al directorio
```

Una vez allí, ingresamos el siguiente comando para crear un repositorio .git:


```
# Creamos el nuevo repositorio
git init
```

Si ya existe un repositorio Git para el proyecto, entonces debemos  introducir el comando *git clone* seguido de la dirección web o de red de ese repositorio para crear una copia de trabajo en el computador:

```
# Clonar un repositorio
git clone https://algunsitioweb/repositorio-git
```

## Revisar el estado del repositorio y agregar nuevos archivos al control de versiones

Uno de los conceptos básicos para utilizar Git es organizar adecuadamente el propio directorio de trabajo.  Esto permite no solo proponer cambios o mejoras  a un proyecto, que posteriormente son aceptados mediante el comando *git commit* (enviar), sino también permite obtener información sobre las actividades de otros usuarios.

Podemos comprobar si nuestra copia de trabajo está actualizada ejecutando este comando:


```
# Comprobar el estado de nuestro repositorio
git status
```
Por lo general, en el caso de los repositorios creados recientemente, o cuando el repositorio principal y la copia de trabajo coinciden por completo, recibiremos un aviso de que no se ha modificado el proyecto (*No commits yet*). Además, Git nos informa de que no hemos compartido nuestros cambios para el próximo commit (*nothing to commit*).

Para añadir un nuevo archivo al control de versiones o para registrar algún cambio para el siguiente *commit*, introducimos el comando *git add* y el nombre de este archivo, que debe encontrarse en nuestro directorio de trabajo. Por ejemplo, para añadir un documento de texto llamado “Test”:


```
# Agregar el archivo Test.txt
git add Test.txt
```
Después, cuando volvamos a comprobar el estado del repositorio, veremos que el documento está a la espera de someterse a la siguiente fase de confirmación de cambios del proyecto, en que estos se aceptarán o no (*Changes to be commited*):

```
$ git status
On branch master

No commits yet

Changes to be commited:
  (use "git rm --cached <file>..." to unstage)
        new file: Test.txt
```

## Confirmar los cambios mediante commit y añadirlos al HEAD

Cualquier cambio que hayamos propuesto incorporar al proyecto, como se ha explicado en el punto anterior, debe confirmarse con *commit* para que se incluya en el *HEAD*. El *HEAD* es un tipo de índice que apunta al último *commit* efectuado en el entorno actual de trabajo Git, tambien se le conoce como “*rama*”.  El comando para hacerlo es el siguiente:

```
# Confirmamos los cambiamos
git commit
```

Después de ejecutar el comando, Git inicia automáticamente el editor que hemos configurado como predeterminado durante la instalación o que el propio sistema abre por defecto. En el documento, podemos añadir un comentario personal sobre el commit planificado, en el que las líneas anotadas se separan por punto y coma y, por lo tanto, no se muestran más adelante. En cuanto cerramos el editor, Git creará el commit:

```
$ git commit
[master (root-commit) c0fdc90] Test
 1 file changed, 0 insertions(+), 0 deletions(-)
 create mode 100644 Test.txt
```
Como podemos ver, después de ejecutar git commit, obtenemos un mensaje que resume el commit: entre corchetes figuran, por un lado, el nombre de la rama del proyecto a la que se transfirieron los cambios (en este caso, *master*, ya que nuestro repositorio de trabajo también es el repositorio principal) y, por otra parte, la suma de comprobación *SHA-1* del *commit* (en este caso, *c0fdc90*). Les siguen el comentario que anotó el propio usuario (aquí, “*Test*”) y algunos datos concretos sobre los cambios.


## Revisar o deshacer commits ejecutados

Una vez aceptados los cambios mediante el comando commit, podemos editar el contenido o eliminarlo por completo en cualquier momento más adelante. 

Por ejemplo, un caso típico sería que al ejecutar *commit* nos hayamos olvidado de algún archivo o configuración importante. En este caso, podemos registrar archivos nuevos o modificados a posteriori mediante el comando *git add* y volver a transferirlos. Para ello, añadimos *--amend* al comando estándar:

```
git commit --amend
```
Si queremos deshacer el último *commit* generado, podemos hacerlo con el siguiente comando de Git:

```
git reset --soft HEAD~1
```
Este comando cancela el *commit* registrado por última vez en el *HEAD*. Los archivos que contiene se restablecen como cambios planificados para el próximo *commit* en el estado del proyecto. Si lo que queremos es eliminar por completo los datos introducidos, ejecutamos el siguiente comando en lugar del anterior:

```
git reset --hard HEAD~1
```

## Mostrar el historial de commits

Aprender a gestionar proyectos con Git es especialmente útil debido a las características básicas de control de versiones que ofrece el sistema . Por ejemplo, una gran ventaja de este programa de código abierto es que siempre puedomes visualizar los últimos cambios que se han realizado en el repositorio. Para ello, puedes utilizar el siguiente comando de Git:

```
# Visualizar la bitacora de cambios
git log
```
De manera predeterminada, el comando *git log* enumera los *commits* generados en orden cronológico inverso: la suma de comprobación *SHA-1*, el autor (nombre y dirección de correo electrónico) y la fecha de cada *commit*.  Además, se muestra un comentario individual que sirve a todos los usuarios como indicador para poder buscar rápidamente las versiones.  Por ejemplo, anteriormente cuando generamos un solo *commit* con el mensaje “Test”. Al ejecutar el comando, se nos muestra el archivo solicitado:

```
$ git log
commit da39a3ee5e6b4b0d3255bfef95601890afd80709 (HEAD -> master)
Author: esvin.molina <3222656594+esvin.molina@users.noreply.github.com>
Date:   Mon Feb 8 10:51:43 2022 +0100

    Test
```
El comando *git log* también puede modificarse utilizando varios parámetros. En la siguiente tabla, se muestran algunas de las posibilidades:

| Parámetros para git log      | Descripción |
| ----------- | ----------- |
| -p      | Muestra también los cambios incluidos en un commit       |
| -2   | Enumera solo los dos últimos commits ejecutados        |
| --stat   | Añade una pequeña estadística a cada registro que muestra qué archivos se han modificado y cuántas líneas se han insertado o eliminado        |
| --pretty   | Cambia el formato de salida con diferentes posibilidades; por ejemplo, --pretty=online enumera todos los commits en una sola línea        |
| --abbrev-commit   | Muestra solo los primeros caracteres de una suma de comprobación SHA-1        |
| --relative-date   | Muestra la fecha de un cambio en formato relativo (por ejemplo, “hace dos semanas”)        |

## Incluir commits en el repositorio principal

Hasta ahora, hemos mostrado cómo guardar los cambios como *commit* en el *HEAD* del directorio local. Para que estos también se incluyan en el repositorio principal, debemos ejecutar el siguiente comando:

```
git push origin master
```
De este modo, Git transfiere automáticamente todos los *commits* ejecutados, que hasta ahora solo habían estado en la copia de trabajo, al directorio principal, que también recibe el nombre de *master*. Si systituimos este nombre por el de otra rama (la del proyecto), los archivos se enviarán directamente allí.

## Crear, eliminar y enumerar etiquetas en Git

Al igual que muchos otros sistemas de control de versiones, Git tiene una función de etiquetado que resalta los elementos seleccionados en el historial del repositorio. Por lo general, estas etiquetas se utilizan para identificar versiones del software, como 1.0, 2.0 o posteriores, y permitir que estas sean fáciles de acceder en todo momento, incluso en el caso de los proyectos grandes.  Git admite dos tipos de etiquetas:

* **Etiquetas anotadas (annotated)**: se guardan como objetos independientes en la base de datos e incluyen su propia suma de comprobación, etiqueta, fecha, nombre y dirección de correo electrónico del creador de la etiqueta, así como la firma opcional GNU Privacy Guard (GPG).
* **Etiquetas no anotadas (lightweight)**: al igual que las ramas, solo sirven de referencia para identificar un *commit*. Este tipo resulta útil si solo necesitamos una etiqueta temporal o si no queremos guardar datos avanzados.

Para crear etiquetas anotadas en Git, ejecutamos el comando *git tag -a* en el *commit* que deseemos. Si también añadimos el parámetro *-m*, podremos anotar entre comillas el mensaje que debe incluir la etiqueta directamente en el símbolo del sistema.  Como ejemplo, el commit “*Test*” que anteriormente hemos generado, los vincularemos a una etiqueta que incluye el mensaje “*example tag*”:

```
git tag -a Test -m “example tag”
```
Para las *etiquetas no anotadas*, procedemos de manera similar: en este caso, solo ejecutamos el comando básico *git tag* en el *commit* deseado y no utilizamos ningún otro parámetro. En el ejemplo que hemos venido trabajando, el comando sería el siguiente:

```
git tag Test
```
En cuanto haya etiquetas en el repositorio, podemos mostrarlas con el mencionado *git tag* y los parámetros opcionales *-lo --list*:

```
git tag
git tag -l
git tag --list
```
Para eliminar una etiqueta del directorio de trabajo local, ejecutamos el comando *git tag -d*. Nuestra referencia a “*Test*” se elimina de la siguiente manera:

```
git tag -d Test
```
Debemos tener cuenta que también es necesario transferir las etiquetas al repositorio principal de forma manual como *commit*. Para ello, introducimos el nombre de la etiqueta y el comando *git push origin*. También puedemos añadir el parámetro *--tags* en vez del nombre de la etiqueta, mediante el cual todas las etiquetas generadas se incluyen en el repositorio:

```
git push origin --tags
```

## Crear, administrar y eliminar ramas

Las ramas ya mencionadas anteriormente no son, en principio, más que versiones de trabajo individuales del repositorio principal, que también se clasifica como una rama llamada *master*. Mediante esta estructura, Git proporciona una base perfecta para desarrollar características y funciones de forma independiente y combinarlas en etapas posteriores, lo que también recibe el nombre de *fusionar* (en inglés, ***merge***).

Crear una nueva rama es fácil: solo debes ejecutar el comando *git branch* y añadir el nombre que queramos poner a la rama. Por ejemplo, puedemos generar una rama de ejemplo llamada *test_branch* de la siguiente manera:

```
git branch test_branch
```
Después, puedemos cambiar a esta rama en cualquier momento mediante el comando *git checkout*:

```
git checkout test_branch
```
Para combinar dos o más ramas, utilizamos el comando *git merge*. Antes, ejecutamos *checkout* para cambiar al directorio que debe incluir la otra rama y, después, ejecutamos allí este comando, incluyendo el nombre de la rama que se fusionará. Nuestra versión de trabajo *test_branch* se puede fusionar con el repositorio principal de la siguiente manera:

```
git checkout master
git merge test_branch
```
Si hemos fusionado varias ramas de trabajo y, por lo tanto, ya no nos hace falta una rama en particular, puedemos eliminarla fácilmente. Para ello, introducimos el comando *git branch -d* y la rama que ya no necesitemos. Eliminaremos *test_branch*, en nuestro ejemplo, mediante el siguiente comando:

```
git branch -d test_branch
```
El único requisito para eliminar una rama es encontrarse en otra distinta en ese momento. Por lo tanto, hemos de cambiar al repositorio principal antes de ejecutar el comando.
