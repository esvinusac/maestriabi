```python
# PROBLEMA No. 1 

# Procedimiento que despliega la representacion binaria del numero decimal especificado.
def decimalABinario(pDecimal):
    vDecimal = pDecimal
    nBinario = 0
    vMultiplicador = 1

    while vDecimal != 0:
        nBinario = nBinario + vDecimal % 2 * vMultiplicador
        vDecimal = vDecimal // 2
        vMultiplicador = vMultiplicador * 10

    print(f"{pDecimal} en decimal a binario es {nBinario}")
  
try:
  numDecimal = int(input('Ingrese un numero entero en base decimal: '))
  
  # validamos que sea un numero entero valido
  if (numDecimal < 0):
    raise ValueError('El numero decimal no debe ser negativo')

  decimalABinario(numDecimal)
  
except ValueError as err:
  print('Los datos no son validos: ',str(err.args[0]))
except Exception as ex:
  print('Hubo un problema: ', ex)  
```

```python
# PROBLEMA No. 2 

# Funcion que devuelve verdadero si el numero especificado es un numero Primo,
# de lo contrario retorna falso.
def esNumPrimo(pNum):
    if pNum < 1:
        return False
    elif pNum == 2:
        return True
    else:
        for i in range(2, pNum):
            if pNum % i == 0:
                return False
        return True 

try:
  num = int(input('Ingrese un numero entero: ')) 
  if esNumPrimo(num):
    print(f"{num} en un numero primo")
  else:
    print(f"{num} no es un numero primo")

except ValueError as err:
  print('Los datos no son validos: ',str(err.args[0]))
except Exception as ex:
  print('Hubo un problema: ', ex)  
```

```python
# PROBLEMA No. 3
import math 

try:
  print('Ingrese lista de valores (separados por coma).  ejemplo: 5,3,2,10 ')
  strValores = input() 

  # convertimos la lista (string) de valores a una lista de enteros
  listaStrValores = strValores.split(',')
  listaValores = [float(valor) for valor in listaStrValores]

  if len(listaValores) == 0:
    print('La lista esta vacia.')
  else:
    # buscamos la desviacion estandar de la lista de valores, calculando 
    # el promedio y la varianza
    promedio = sum(listaValores) / len(listaValores)
    varianza = sum((valor-promedio)**2 for valor in listaValores) / len(listaValores)
    desviacionEstandar =  math.sqrt(varianza)
    print(f"La desviacion estandar de la lista es: {str(desviacionEstandar)}")

except ValueError as err:
  print('Los datos no son validos: ',str(err.args[0]))
except Exception as ex:
  print('Hubo un problema: ', ex)  
```

```python
# PROBLEMA No. 4 

# procedimiento de ordenamiento por el algoritmo de burbuja
def burbuja(pArreglo):
  vArreglo = pArreglo[:] # copiamos el parametro de entrada para no modificarlo
  for i in range(1,len(vArreglo)):
      for j in range(0,len(vArreglo)-i):
          if(vArreglo[j+1] < vArreglo[j]):
              aux=vArreglo[j]
              vArreglo[j]=vArreglo[j+1]
              vArreglo[j+1]=aux
  return vArreglo

# programa principal
def main():
  
  # este es el arreglo de entrada a ordenar, puede ser modificado aqui
  info=[6,5,3,1,8,7,2,4,] 

  print('Sin ordenar: ', info)
  infoOrdenada = burbuja(info)
  print('Odenado: ', infoOrdenada)

if __name__ == "__main__":
    main()  
```

```python
# PROBLEMA No. 5 

# Funcion que convierte una tupla a una cadena de caracteres
def tuplaToStr(pTupla):
  cadena = ''
  for valor in pTupla:
    cadena += str(valor)
  return cadena

try:
  # esta es la tupla de entrada, puede ser modificada aqui
  tupla = (10, 20, 40, 5, 70)
  
  print(f'Tupla original: {tupla}')
  print(f'Tupla a String: {tuplaToStr(tupla)}')

except ValueError as err:
  print('Los datos no son validos: ',str(err.args[0]))
except Exception as ex:
  print('Hubo un problema: ', ex)  
```

```python
# PROBLEMA No. 6

# Funcion que elimina las tuplas vacias de un listado de tuplas
def eliminarTuplasVacias(pTuplas):
    vTuplas = [tupla for tupla in pTuplas if tupla]
    return vTuplas

try:
  # este es el listado de tuplas de entrada, pueden ser modificadas aqui
  tuplas =  [(), (), ('X',), ('a', 'b'), ('a', 'b', 'c'), ('d')]
  
  print(f"Tuplas originales: {str(tuplas)}") 
  print(f"Listado sin tuplas vacias: {str(eliminarTuplasVacias(tuplas))}")
  
except ValueError as err:
  print('Los datos no son validos: ',str(err.args[0]))
except Exception as ex:
  print('Hubo un problema: ', ex)  
```

```python
# PROBLEMA No. 7

# Funcion que regresa una tupla con los promedios de cada una de las 
# tuplas que se enceuntra en el listado especificado
def promedioTuplas(pTuplas):
  lPromedios=[]  #contendra el listado de promedios
  [lPromedios.append(sum(tupla)/len(tupla)) for tupla in pTuplas]
  tPromedios = tuple(lPromedios)
  return tPromedios

try:
  # este es el listado de tuplas de entrada, pueden ser modificadas aqui
  tuplas = ((10, 10, 10, 12), (30, 45, 56, 45), (81, 80, 39, 32))

  print(f"Tuplas originales : {str(tuplas)}") 
  print(f"Promedios: {str(promedioTuplas(tuplas))}")

except ValueError as err:
  print('Los datos no son validos: ',str(err.args[0]))
except Exception as ex:
  print('Hubo un problema: ', ex)  
```
